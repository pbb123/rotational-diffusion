```python
import numpy as np
import matplotlib.pyplot as plt
import scipy as scp
```

```python
data = np.loadtxt("out.txt",delimiter=" ")
data /= 1e4
```

```python
cx = scp.signal.correlate(data[:,0],data[:,0],mode="same")
cy = scp.signal.correlate(data[:,1],data[:,1],mode="same")
cz = scp.signal.correlate(data[:,2],data[:,2],mode="same")
```

```python
c=cx+cy+cz
```

```python
x=np.linspace(0,1,1000)
plt.plot(1e-6*np.arange(len(c)),c)
plt.plot(x,np.exp(-3*np.abs(x-0.5)))
```
